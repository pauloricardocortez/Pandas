```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Files/source of capital.csv')
columns = ['Industry', 'Gender_ethnicity', 'Source', 'Total_firms', 'Total_sales']
df.columns = columns
df = df.drop(columns = 'Total_sales')
df.head()
```

```python
#Check data types
df.dtypes
```

```python
#Check the column
df['Total_firms'].unique()
```

```python
#Convert to int
df['Total_firms'] = df['Total_firms'].str.replace(r"\(.*\)","")
df['Total_firms'] = pd.to_numeric(df['Total_firms'])
```

```python
#Check if col was converted
df.dtypes
```

```python
#Remove all firms?
df_grouped_source = df.groupby('Source').sum()
df_grouped_source = df_grouped_source.sort_values('Total_firms', ascending = False)
df_grouped_source['Percentage'] = round(df_grouped_source['Total_firms']/df_grouped_source.iloc[0,0]*100,2)
df_grouped_source
```
