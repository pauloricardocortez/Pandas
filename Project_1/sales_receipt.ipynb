```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Files/business activity.csv')
columns = ['Industry', 'Gender_ethnicity', 'Sales_receipt', 'Operations_time', 'Total_firms', 'Total_sales']
df.columns = columns
df = df.drop(columns = 'Total_sales')
df.head()
```

```python
#Check data types
df.dtypes
```

```python
#Group by sales
df_grouped_sales = df.groupby('Sales_receipt').sum()
df_grouped_sales = df_grouped_sales.sort_values('Total_firms', ascending = False)
df_grouped_sales['Percentage'] = round(df_grouped_sales['Total_firms']/df_grouped_sales.iloc[0,0]*100,2)
df_grouped_sales
```

```python
#Group by operations time
#Remove all firms?
df_grouped_ops = df.groupby('Operations_time').sum()
df_grouped_ops = df_grouped_ops.sort_values('Total_firms', ascending = False)
df_grouped_ops['Percentage'] = round(df_grouped_ops['Total_firms']/df_grouped_ops.iloc[0,0]*100,2)
df_grouped_ops
```

```python
#Sales by industry
df_industry = df.set_index(['Industry', 'Sales_receipt'])
df_industry
```
