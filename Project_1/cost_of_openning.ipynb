```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Files/amount to open.csv')
columns = ['Industry', 'Gender_ethnicity', 'Capital_needed', 'Total_firms', 'Total_sales']
df.columns = columns
df = df.drop(columns = 'Total_sales')
df.head()
```

```python
#Check data types
df.dtypes
```

```python
df_grouped_gender = df.groupby('Gender_ethnicity').sum()
df_grouped_gender = df_grouped_gender.sort_values('Total_firms', ascending = False)
df_grouped_gender['Percentage'] = round(df_grouped_gender['Total_firms']/df_grouped_gender.iloc[0,0]*100,2)
df_grouped_gender.head()
```

```python
#Remove All firms?
df_grouped_capital = df.groupby('Capital_needed').sum()
df_grouped_capital = df_grouped_capital.sort_values('Total_firms', ascending = False)
df_grouped_capital['Percentage'] = round(df_grouped_capital['Total_firms']/df_grouped_capital.iloc[0,0]*100,2)
df_grouped_capital
```
