```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Files/ceased operations.csv')
columns = ['Industry', 'Gender_ethnicity', 'Ceased_reasons', 'Total_firms', 'Total_sales']
df.columns = columns
df = df.drop(columns = 'Total_sales')
df.head()
```

```python
#Check data types
df.dtypes
```

```python
#Remove all firms?
df_grouped_ceased = df.groupby('Ceased_reasons').sum()
df_grouped_ceased = df_grouped_ceased.sort_values('Total_firms', ascending = False)
df_grouped_ceased['Percentage'] = round(df_grouped_ceased['Total_firms']/df_grouped_ceased.iloc[0,0]*100,2)
df_grouped_ceased
```
