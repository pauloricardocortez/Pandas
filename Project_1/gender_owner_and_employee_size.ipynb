```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('number_of_employees.csv')
df = df.drop(columns = 'Meaning of Types of workers used code')
columns = ['Industry', 'Gender_ethnicity', 'Size', 'Total']
df.columns = columns
df.head()
```

```python
df.dtypes
```

```python
new_df = df.loc[(df['Industry'] == 'Total for all sectors')]
new_df = new_df.drop(columns = 'Industry')
new_df = new_df.set_index('Gender_ethnicity')
new_df.head()
```

```python
gender_group = new_df[(new_df['Size'] == 'All firms')]
gender_group = gender_group.drop(columns = 'Size')
gender_group = gender_group.groupby('Gender_ethnicity').sum()
gender_group['Percentage'] = round((gender_group['Total']/gender_group.iloc[0,0])*100,2)
gender_group
```

```python
alldf = new_df.loc['All firms']
alldf = alldf.groupby('Size').sum()
alldf['Percentage'] = round((alldf['Total'] / alldf.iloc[0,0])*100,2)
alldf
```

```python
femaledf = new_df.loc['Female-owned']
femaledf = femaledf.groupby('Size').sum()
femaledf['Percentage'] = round((femaledf['Total'] / femaledf.iloc[0,0])*100,2)
femaledf
```

```python
maledf = new_df.loc['Male-owned']
maledf = maledf.groupby('Size').sum()
maledf['Percentage'] = round((maledf['Total'] / maledf.iloc[0,0])*100,2)
maledf
```
