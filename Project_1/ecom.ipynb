```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Files/ecommerce.csv')
columns = ['Industry', 'Gender_ethnicity', 'Ecommerce_sales', 'Total_firms', 'Total_sales']
df.columns = columns
df = df.drop(columns = 'Total_sales')
df.head()
```

```python
#Remove all firms?
df_grouped_ecom = df.groupby('Ecommerce_sales').sum()
df_grouped_ecom = df_grouped_ecom.sort_values('Total_firms', ascending = False)
df_grouped_ecom['Percentage'] = round(df_grouped_ecom['Total_firms']/df_grouped_ecom.iloc[0,0]*100,2)
df_grouped_ecom
```
